```python
import yaml, requests
from io import BytesIO
from PIL import Image
from IPython.display import display
from datetime import datetime
import torch
from diffusers import AutoPipelineForText2Image

from google import genai
from google.genai import types

# Ë®≠ÂÆöË™≠„ÅøËæº„Åø
with open("config.yml", "r") as f:
    cfg = yaml.safe_load(f)

API_KEY     = cfg["gemini_api_key"]
TEXT_MODEL  = cfg.get("text_model", "gemini-2.5-flash")
IMAGE_MODEL = cfg.get("image_model", "imagen-3.0-generate-002")
STYLE_HINT  = cfg.get("style_hint", "cute mascot, soft pastel, simple bg, no text")

client = genai.Client(api_key=API_KEY)

# 1) „Ç∏„Éß„Éº„ÇØÂèñÂæó
def fetch_random_dad_joke(ua="Colab-Gemini-Demo/1.0") -> str:
    url = "https://icanhazdadjoke.com/"
    headers = {"Accept": "application/json", "User-Agent": ua}  # ‚Üê „Éò„ÉÉ„ÉÄ„ÅØASCII„ÅÆ„Åø
    r = requests.get(url, headers=headers, timeout=10)
    r.raise_for_status()
    return r.json().get("joke", "No joke found!")

# 2) Êó•Êú¨Ë™û„Ç≠„É£„Éó„Ç∑„Éß„É≥ÔºàÁü≠ÊñáÔºâ  ‚Üê FIX: contents „Çí types.Content / types.Part „ÅßÊßãÈÄ†Âåñ
def make_caption_ja(joke: str) -> str:
    prompt = (
        "Ê¨°„ÅÆËã±Ë™û„Ç∏„Éß„Éº„ÇØ„Å´„ÄÅÁü≠„Åè„Å¶ÁßÄÈÄ∏„Å™Êó•Êú¨Ë™û„Ç≠„É£„Éó„Ç∑„Éß„É≥„Çí1„Å§„Å†„Åë‰ΩúÊàê„ÄÇ"
        "40Â≠ó‰ª•ÂÜÖ„ÄÅÁµµÊñáÂ≠ó„ÅØ1ÔΩû2ÂÄã„ÄÅË™ûÂ∞æ„ÅØËªΩ„ÇÅ„ÄÇÂá∫Âäõ„ÅØ„Ç≠„É£„Éó„Ç∑„Éß„É≥„ÅÆ„Åø„ÄÇ\n\n"
        f"{joke}"
    )
    res = client.models.generate_content(
        model=TEXT_MODEL,
        contents=[  # ‚Üê list[types.Content]
            types.Content(
                role="user",
                parts=[types.Part.from_text(text=prompt)]
            )
        ],
    )
    return (res.text or "").strip()

# 3) ÁîªÂÉèÁî®„Éó„É≠„É≥„Éó„ÉàÔºàËã±Ë™û„Å∏Êï¥ÂΩ¢Ôºâ ‚Üê FIX: ÂêåÊßò„Å´ÊßãÈÄ†Âåñ„Åó„Å¶Ê∏°„Åô
def build_image_prompt_en(joke: str, style_hint: str = STYLE_HINT) -> str:
    sys_text = (
        "You are an art director. Convert the joke into a concise English text-to-image prompt "
        "for a cute illustration. Include subject, scene, facial expression, small props, composition, "
        "color tone, and style. Do NOT include any text to be rendered in the image. 50‚Äì120 words."
    )
    user_text = f"Joke:\n{joke}\n\nPreferred style: {style_hint}"

    res = client.models.generate_content(
        model=TEXT_MODEL,
        contents=[
            types.Content(
                role="user",
                parts=[
                    types.Part.from_text(text=sys_text),   # ‚Üê text= „Çí‰ªò„Åë„Çã
                    types.Part.from_text(text=user_text),  # ‚Üê text= „Çí‰ªò„Åë„Çã
                ],
            )
        ],
    )


    return (res.text or "").strip()

# 4) ÁîªÂÉèÁîüÊàêÔºàImagen 3Ôºâ
def make_image(prompt_en: str, n=1):
    cfg_img = types.GenerateImagesConfig(number_of_images=n)
    res = client.models.generate_images(
        model=IMAGE_MODEL,
        prompt=prompt_en,
        config=cfg_img,
    )
    imgs = []
    for gi in res.generated_images:
        imgs.append(Image.open(BytesIO(gi.image.image_bytes)))
    return imgs

# ÂÆüË°å
joke = fetch_random_dad_joke()
caption = make_caption_ja(joke)
img_prompt = build_image_prompt_en(joke)

print("„ÄêJoke„Äë", joke)
print("„ÄêCaption(ja)„Äë", caption)
print("„ÄêImage Prompt(en)„Äë", (img_prompt[:200] + "‚Ä¶") if len(img_prompt) > 200 else img_prompt)

# ÁîªÂÉèÁîüÊàê‚ÜíË°®Á§∫‚Üí‰øùÂ≠ò
_sd_pipeline = None
def make_image_free(prompt_en: str, n=1, seed: int | None = 1234):
    global _sd_pipeline
    if _sd_pipeline is None:
        _sd_pipeline = AutoPipelineForText2Image.from_pretrained(
            "stabilityai/sdxl-turbo", torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32
        )
        device = "cuda" if torch.cuda.is_available() else "cpu"
        _sd_pipeline = _sd_pipeline.to(device)

    g = None
    if seed is not None:
        g = torch.Generator(device=_sd_pipeline.device).manual_seed(seed)

    # SDXL-Turbo „ÅØ‰Ωé„Çπ„ÉÜ„ÉÉ„Éó„Åß„ÇÇÂá∫„ÇÑ„Åô„ÅÑ
    images = _sd_pipeline(
        prompt=prompt_en,
        num_inference_steps=4,
        guidance_scale=0.0,   # turboÁ≥ª„ÅØ 0.0„Äú1.5 Êé®Â•®
        generator=g
    ).images

    return images[:n]
```

```python
prompt = img_prompt

import torch
from diffusers import AutoPipelineForText2Image
from IPython.display import display

# ÈÄü„Åè„Å¶ÁÑ°Êñô„Åß„Åù„Åì„Åù„ÅìÁ∂∫È∫óÔºöSDXL Turbo
dtype = torch.float16 if torch.cuda.is_available() else torch.float32
pipe = AutoPipelineForText2Image.from_pretrained(
    "stabilityai/sdxl-turbo", torch_dtype=dtype
).to("cuda" if torch.cuda.is_available() else "cpu")

seed = 1234  # ÂÜçÁèæÁî®„ÄÇÊØéÂõûÂ§â„Åà„Åü„ÅÑ„Å™„Çâ None „Å´
g = torch.Generator(device=pipe.device).manual_seed(seed)

images = pipe(
    prompt=prompt,
    num_inference_steps=4,   # 2„Äú6„ÅßÈ´òÈÄü„ÄÇÂìÅË≥™‰∏ä„Åí„Åü„ÅÑ„Å™„Çâ8„Äú12
    guidance_scale=0.0,      # turbo„ÅØ0.0„Äú1.5„Åå„Åä„Åô„Åô„ÇÅ
    generator=g
).images

paths = []
for i, im in enumerate(images, 1):
    display(im)
    out = f"/content/mascot_{i}.png"
    im.save(out, "PNG")
    paths.append(out)
print("saved:", paths)
```

```python
# === ADK „Å´ SDXL Turbo ÁîªÂÉèÁîüÊàê„Çí„ÉÑ„Éº„É´ÁôªÈå≤„Åô„ÇãÊúÄÂ∞èÂÆüË£Ö ===
import os, time, random, torch
from io import BytesIO
from PIL import Image
from IPython.display import display
from diffusers import AutoPipelineForText2Image

# ADK
from google.adk.agents.llm_agent import LlmAgent
from google.adk.runners import Runner
from google.adk.sessions import InMemorySessionService

# „Åô„Åß„Å´ Gemini „ÅÆ TEXT_MODEL „Çí‰Ωø„Å£„Å¶„ÅÑ„Çã„Å™„ÇâÊµÅÁî®„ÄÇÊú™ÂÆöÁæ©„Å™„Çâ„ÉÄ„Éü„Éº„Åß„ÇÇÂèØ„ÄÇ
try:
    TEXT_MODEL  # noqa
except NameError:
    TEXT_MODEL = "gemini-2.5-flash"

# ---- SDXL Turbo „Éë„Ç§„Éó„É©„Ç§„É≥ÔºàÈÅÖÂª∂ÂàùÊúüÂåñ & ‰Ωø„ÅÑÂõû„ÅóÔºâ----
_sdxl_pipe = None
_device = "cuda" if torch.cuda.is_available() else "cpu"
_dtype = torch.float16 if torch.cuda.is_available() else torch.float32

def _ensure_sdxl():
    global _sdxl_pipe
    if _sdxl_pipe is None:
        _sdxl_pipe = AutoPipelineForText2Image.from_pretrained(
            "stabilityai/sdxl-turbo", torch_dtype=_dtype
        ).to(_device)
    return _sdxl_pipe

# ---- ADK „ÉÑ„Éº„É´ÔºöÁîªÂÉèÁîüÊàêÔºàDiffusers / ÁÑ°ÊñôÔºâ----
def generate_image_sdxl(
    prompt: str,
    n: int = 1,
    steps: int = 4,
    guidance: float = 0.0,
    seed: int | None = 1234,
    display_in_colab: bool = True,
) -> dict:
    """
    SDXL-Turbo „ÅßÁîªÂÉè„ÇíÁîüÊàê„Åó„Å¶‰øùÂ≠ò„Éë„Çπ„ÇíËøî„Åô„ÉÑ„Éº„É´„ÄÇ
    - prompt: Ëã±Ë™û„ÅÆÁîªÂÉè„Éó„É≠„É≥„Éó„Éà
    - n: ÊûöÊï∞
    - steps: ÂèçÂæ©„Çπ„ÉÜ„ÉÉ„ÉóÔºàTurbo „ÅØ 2„Äú6 „ÅßÈ´òÈÄü„ÄÅÂìÅË≥™‚Üë„Å™„Çâ 8„Äú12Ôºâ
    - guidance: 0.0„Äú1.5 Êé®Â•®ÔºàTurboÔºâ
    - seed: ÂÜçÁèæÁî®„ÄÇNone „Å™„ÇâÈÉΩÂ∫¶„É©„É≥„ÉÄ„É†
    - display_in_colab: True „Å™„Çâ display() „ÅßË°®Á§∫
    return: {"paths": [...]}  ÁîüÊàêÁîªÂÉè„ÅÆ‰øùÂ≠ò„Éë„ÇπÈÖçÂàó
    """
    pipe = _ensure_sdxl()

    paths = []
    base = f"/content/mascot_{int(time.time())}"
    for i in range(n):
        g = None
        if seed is not None:
            # Ë§áÊï∞Êûö„Åß„ÇÇÂ§âÂåñ„ÅåÂá∫„Çã„Çà„ÅÜ„Å´„Ç™„Éï„Çª„ÉÉ„Éà
            g = torch.Generator(device=pipe.device).manual_seed(seed + i)

        out = pipe(
            prompt=prompt,
            num_inference_steps=steps,
            guidance_scale=guidance,
            generator=g,
        )
        img: Image.Image = out.images[0]
        path = f"{base}_{i+1}.png"
        img.save(path, "PNG")
        paths.append(path)
        if display_in_colab:
            display(img)

    return {"paths": paths}

# ---- ADK „Ç®„Éº„Ç∏„Çß„É≥„ÉàÂÆöÁæ© ----
agent = LlmAgent(
    name="ImageGenAgent",
    instruction=(
        "You are an assistant that generates images using the SDXL Turbo tool. "
        "When the user provides a prompt (or mentions 'img_prompt'), "
        "call the tool 'generate_image_sdxl' with that prompt and return only the saved file paths."
    ),
    model=TEXT_MODEL,
    tools=[generate_image_sdxl],
)

runner = Runner(
   app_name="JokeImageApp",                # ‚Üê „Åì„Çå„ÅåÂøÖÈ†à
    agent=agent,
    session_service=InMemorySessionService()
)
# ---- ÂÆüË°å„É©„ÉÉ„ÉëÔºàimg_prompt „ÇíËá™Âãï„ÅßÊ∏°„ÅôÔºâ----
def adk_generate_from_img_prompt(
    prompt_var_name: str = "img_prompt",
    n: int = 1,
    steps: int = 4,
    guidance: float = 0.0,
    seed: int | None = 1234,
):
    if prompt_var_name not in globals():
        raise RuntimeError(f"{prompt_var_name} „ÅåË¶ã„Å§„Åã„Çä„Åæ„Åõ„Çì„ÄÇÂÖà„Å´„Éó„É≠„É≥„Éó„Éà„Çí‰ΩúÊàê„Åó„Å¶„Åè„Å†„Åï„ÅÑ„ÄÇ")
    p = globals()[prompt_var_name]

    msg = (
        "Use the tool 'generate_image_sdxl' with the following arguments and return only the file paths.\n"
        f"prompt: '''{p}'''\n"
        f"n: {n}\nsteps: {steps}\nguidance: {guidance}\nseed: {seed}\n"
    )

    # --- ADK„ÅÆÁâàÂ∑Æ„ÇíÂê∏Âèé„Åô„ÇãÂëº„Å≥Âá∫„Åó ---
    try:
        # „Åæ„Åö„ÅØ "prompt=" „Åß
        result = runner.run(prompt=msg)
    except TypeError:
        try:
            # ‰∏ÄÈÉ®Áâà„Åß„ÅØ "input_text=" / "query=" „Å™„Å©„ÅÆÂêçÂâç
            result = runner.run(input_text=msg)
        except TypeError:
            try:
                result = runner.run(query=msg)
            except TypeError:
                # „Å©„ÅÜ„Åó„Å¶„ÇÇÂêà„Çè„Å™„ÅÑÂ†¥Âêà„ÅØ„ÄÅ„ÉÑ„Éº„É´„ÇíÁõ¥Êé•Âëº„Çì„ÅßÁµêÊûú„ÇíËøî„ÅôÔºà„Éï„Ç©„Éº„É´„Éê„ÉÉ„ÇØÔºâ
                out = generate_image_sdxl(prompt=p, n=n, steps=steps, guidance=guidance, seed=seed)
                print(out["paths"])
                return

    print(getattr(result, "text", str(result)))
```

```python
# === ‰∏âÊäû„ÇØ„Ç§„Ç∫ÔºöÁîªÂÉè + „Ç∏„Éß„Éº„ÇØ‰∏âÊäûÔºàÊ≠£Ëß£=ÁîªÂÉè„Å´‰Ωø„Å£„Åü„Ç∏„Éß„Éº„ÇØÔºâ ===
import os, random, yaml, requests, ipywidgets as w
from IPython.display import display, clear_output
from PIL import Image
from io import BytesIO

# Gemini (google-genai) „Çí‰Ωø„ÅÜ„ÄÇÊú™ÂàùÊúüÂåñ„Å™„Çâ config.yml „Åã„ÇâË™≠„ÇÄ
try:
    client  # noqa
    TEXT_MODEL  # noqa
except NameError:
    from google import genai
    with open("config.yml", "r") as f:
        _cfg = yaml.safe_load(f)
    API_KEY    = _cfg["gemini_api_key"]
    TEXT_MODEL = _cfg.get("text_model", "gemini-2.5-flash")
    client     = genai.Client(api_key=API_KEY)

from google.genai import types

# ---- „É¶„Éº„ÉÜ„Ç£„É™„ÉÜ„Ç£ ----
HEADERS = {"Accept": "application/json", "User-Agent": "Colab-quiz/1.0"}

def fetch_random_joke_en() -> str:
    r = requests.get("https://icanhazdadjoke.com/", headers=HEADERS, timeout=10)
    r.raise_for_status()
    return r.json().get("joke", "")

def translate_en_to_ja(text_en: str) -> str:
    """Gemini„ÅßËã±‚ÜíÊó•„ÅÆËá™ÁÑ∂„Å™Ë®≥Ôºà1Ë°åÔºâ"""
    prompt = f"Ê¨°„ÅÆËã±Ë™û„Ç∏„Éß„Éº„ÇØ„ÇíËá™ÁÑ∂„Å™Êó•Êú¨Ë™û„Å´1Ë°å„ÅßÁøªË®≥„Åó„Å¶„Åè„Å†„Åï„ÅÑ„ÄÇÂºïÁî®Á¨¶„ÅØ‰ªò„Åë„Åö„ÄÅÂéüÊñá„ÅÆ„Éã„É•„Ç¢„É≥„Çπ„ÅØ‰øù„Å£„Å¶„Åè„Å†„Åï„ÅÑ„ÄÇ\n\n{text_en}"
    res = client.models.generate_content(
        model=TEXT_MODEL,
        contents=[
            types.Content(
                role="user",
                parts=[types.Part.from_text(text=prompt)]
            )
        ],
    )
    return (res.text or "").strip()

def pick_unique_distractors(correct_en: str, k: int = 2) -> list[str]:
    """Ê≠£Ëß£„Å®ÈáçË§á„Åó„Å™„ÅÑ„É©„É≥„ÉÄ„É†„Ç∏„Éß„Éº„ÇØ„ÇíkÂÄãÈõÜ„ÇÅ„Çã"""
    s = set([correct_en.strip()])
    out = []
    for _ in range(20):  # ÊúÄÂ§ß20Âõû„Éà„É©„Ç§
        if len(out) >= k:
            break
        j = fetch_random_joke_en().strip()
        if j and j not in s:
            s.add(j)
            out.append(j)
    if len(out) < k:
        # „Åæ„Çå„Å´ÈáçË§áÁ∂ö„Åç„ÅÆ‰øùÈô∫
        while len(out) < k:
            out.append("Why did the math book look sad? Because it had too many problems.")
    return out[:k]

# ---- ÁîªÂÉè„ÅÆÂèñÂæóÔºà„ÅÇ„Å™„Åü„ÅÆ„Éé„Éº„Éà„ÅÆÂ§âÊï∞„Åã„ÇâÊãæ„ÅÜÔºâ----
# 1) Êó¢Â≠ò„ÅÆ 'images' (PIL.Image „ÅÆ„É™„Çπ„Éà) „Åå„ÅÇ„Çå„Å∞„Åù„Çå„ÇíÂÑ™ÂÖà
# 2) Ê¨°„Å´ 'paths' (‰øùÂ≠ò„Éï„Ç°„Ç§„É´„Éë„Çπ„ÅÆ„É™„Çπ„Éà) „Åå„ÅÇ„Çå„Å∞ÊúÄÂæå„ÅÆ1Êûö„ÇíË°®Á§∫
# 3) „Å©„Çå„ÇÇÁÑ°„Åë„Çå„Å∞„Çπ„Ç≠„ÉÉ„Éó
quiz_image = None
if "images" in globals() and images:
    quiz_image = images[0]
elif "paths" in globals() and paths:
    fp = paths[-1]
    if os.path.exists(fp):
        quiz_image = Image.open(fp)

# ---- Ê≠£Ëß£„Ç∏„Éß„Éº„ÇØÔºàËã±Ë™ûÔºâ„ÅØ„ÄÅ„ÅÇ„Å™„Åü„ÅÆ„Éé„Éº„Éà„ÅÆ 'joke' Â§âÊï∞„ÇíÂà©Áî®„ÄÇÁÑ°„Åë„Çå„Å∞Êñ∞Ë¶èÂèñÂæó ----
try:
    correct_en = joke  # Êó¢Â≠ò„ÅÆËã±Ë™û„Ç∏„Éß„Éº„ÇØÔºàÁîªÂÉè„ÅÆÂÖÉÔºâ
except NameError:
    correct_en = fetch_random_joke_en()

correct_ja = translate_en_to_ja(correct_en)

# ---- Ë™§Á≠îÔºàËã±Ë™ûÔºâ„ÇíAPI„Åã„Çâ ‚Üí Êó•Êú¨Ë™û„Å´Ë®≥„Åô ----
distractors_en = pick_unique_distractors(correct_en, k=2)
distractors_ja = [translate_en_to_ja(x) for x in distractors_en]

# ---- ‰∏âÊäû„Éá„Éº„Çø‰ΩúÊàêÔºà„Ç∑„É£„ÉÉ„Éï„É´Ôºâ----
# ---- ‰∏âÊäû„Éá„Éº„Çø‰ΩúÊàêÔºàËã±Ë™ûË°®Á§∫Ôºâ----
# „Åì„Åì„Åæ„Åß„Åß correct_en, distractors_en „ÅåÊèÉ„Å£„Å¶„ÅÑ„ÇãÂâçÊèê
options = [
    {"label": "A", "text": correct_en,        "is_correct": True},
    {"label": "B", "text": distractors_en[0], "is_correct": False},
    {"label": "C", "text": distractors_en[1], "is_correct": False},
]
random.shuffle(options)  # ‰∏¶„Å≥„Çí„É©„É≥„ÉÄ„É†„Å´

# ---- UI ----
out = w.Output()

if 'quiz_image' in globals() and quiz_image is not None:
    display(quiz_image)

print("Which joke matches this image? (English options)")

rb = w.RadioButtons(
    options=[f"{opt['label']}. {opt['text']}" for opt in options],
    description="Pick:",
    layout=w.Layout(width="100%"),
)
btn = w.Button(description="Submit", button_style="primary")
show_ja = w.Checkbox(value=False, description="Show Japanese translation after grading")
display(rb, show_ja, btn, out)

def on_submit(_):
    with out:
        clear_output()
        selected_text = rb.value
        chosen = None
        for opt in options:
            if selected_text.startswith(opt["label"] + "."):
                chosen = opt
                break
        if chosen is None:
            print("Please choose an option.")
            return

        if chosen["is_correct"]:
            print("‚úÖ Correct!")
        else:
            print("‚ùå Incorrect.")

        # Ê≠£Ëß£„ÅÆËã±Ë™û„ÇíË°®Á§∫
        print("\n„ÄêCorrect (EN)„Äë", correct_en)

        # ‰ªªÊÑèÔºöÊó•Êú¨Ë™ûË®≥„ÇÇË¶ã„Åü„ÅÑÂ†¥ÂêàÔºàÊó¢„Å´ correct_ja „Åå„ÅÇ„Çå„Å∞„Åù„Çå„Çí„ÄÅÁÑ°„Åë„Çå„Å∞„Åù„ÅÆÂ†¥„ÅßÁøªË®≥Ôºâ
        if show_ja.value:
            try:
                correct_ja  # Êó¢„Å´Ë®àÁÆóÊ∏à„Åø„ÅãÔºü
            except NameError:
                # ÁøªË®≥Èñ¢Êï∞„ÅåÂÆöÁæ©Ê∏à„Åø„Å™„ÇâÂÜçÂà©Áî®„ÄÅÁÑ°„Åë„Çå„Å∞„Çπ„Ç≠„ÉÉ„Éó
                if 'translate_en_to_ja' in globals():
                    _ja = translate_en_to_ja(correct_en)
                    print("„ÄêCorrect (JA)„Äë", _ja)
                else:
                    print("„ÄêCorrect (JA)„Äë (translator function not found)")
        # ÔºàÂøÖË¶Å„Å™„ÇâË™§Á≠î„ÅÆËã±Ë™û„ÇíË°®Á§∫„Åô„ÇãÂá¶ÁêÜ„ÇÇËøΩË®òÂèØËÉΩÔºâ

btn.on_click(on_submit)

```
