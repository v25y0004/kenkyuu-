```python
import yaml, requests
from io import BytesIO
from PIL import Image
from IPython.display import display
from datetime import datetime
import torch
from diffusers import AutoPipelineForText2Image

from google import genai
from google.genai import types

# 設定読み込み
with open("config.yml", "r") as f:
    cfg = yaml.safe_load(f)

API_KEY     = cfg["gemini_api_key"]
TEXT_MODEL  = cfg.get("text_model", "gemini-2.5-flash")
IMAGE_MODEL = cfg.get("image_model", "imagen-3.0-generate-002")
STYLE_HINT  = cfg.get("style_hint", "cute mascot, soft pastel, simple bg, no text")

client = genai.Client(api_key=API_KEY)

# 1) ジョーク取得
def fetch_random_dad_joke(ua="Colab-Gemini-Demo/1.0") -> str:
    url = "https://icanhazdadjoke.com/"
    headers = {"Accept": "application/json", "User-Agent": ua}  # ← ヘッダはASCIIのみ
    r = requests.get(url, headers=headers, timeout=10)
    r.raise_for_status()
    return r.json().get("joke", "No joke found!")

# 2) 日本語キャプション（短文）  ← FIX: contents を types.Content / types.Part で構造化
def make_caption_ja(joke: str) -> str:
    prompt = (
        "次の英語ジョークに、短くて秀逸な日本語キャプションを1つだけ作成。"
        "40字以内、絵文字は1～2個、語尾は軽め。出力はキャプションのみ。\n\n"
        f"{joke}"
    )
    res = client.models.generate_content(
        model=TEXT_MODEL,
        contents=[  # ← list[types.Content]
            types.Content(
                role="user",
                parts=[types.Part.from_text(text=prompt)]
            )
        ],
    )
    return (res.text or "").strip()

# 3) 画像用プロンプト（英語へ整形） ← FIX: 同様に構造化して渡す
def build_image_prompt_en(joke: str, style_hint: str = STYLE_HINT) -> str:
    sys_text = (
        "You are an art director. Convert the joke into a concise English text-to-image prompt "
        "for a cute illustration. Include subject, scene, facial expression, small props, composition, "
        "color tone, and style. Do NOT include any text to be rendered in the image. 50–120 words."
    )
    user_text = f"Joke:\n{joke}\n\nPreferred style: {style_hint}"

    res = client.models.generate_content(
        model=TEXT_MODEL,
        contents=[
            types.Content(
                role="user",
                parts=[
                    types.Part.from_text(text=sys_text),   # ← text= を付ける
                    types.Part.from_text(text=user_text),  # ← text= を付ける
                ],
            )
        ],
    )


    return (res.text or "").strip()

# 4) 画像生成（Imagen 3）
def make_image(prompt_en: str, n=1):
    cfg_img = types.GenerateImagesConfig(number_of_images=n)
    res = client.models.generate_images(
        model=IMAGE_MODEL,
        prompt=prompt_en,
        config=cfg_img,
    )
    imgs = []
    for gi in res.generated_images:
        imgs.append(Image.open(BytesIO(gi.image.image_bytes)))
    return imgs

# 実行
joke = fetch_random_dad_joke()
caption = make_caption_ja(joke)
img_prompt = build_image_prompt_en(joke)

print("【Joke】", joke)
print("【Caption(ja)】", caption)
print("【Image Prompt(en)】", (img_prompt[:200] + "…") if len(img_prompt) > 200 else img_prompt)

# 画像生成→表示→保存
_sd_pipeline = None
def make_image_free(prompt_en: str, n=1, seed: int | None = 1234):
    global _sd_pipeline
    if _sd_pipeline is None:
        _sd_pipeline = AutoPipelineForText2Image.from_pretrained(
            "stabilityai/sdxl-turbo", torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32
        )
        device = "cuda" if torch.cuda.is_available() else "cpu"
        _sd_pipeline = _sd_pipeline.to(device)

    g = None
    if seed is not None:
        g = torch.Generator(device=_sd_pipeline.device).manual_seed(seed)

    # SDXL-Turbo は低ステップでも出やすい
    images = _sd_pipeline(
        prompt=prompt_en,
        num_inference_steps=4,
        guidance_scale=0.0,   # turbo系は 0.0〜1.5 推奨
        generator=g
    ).images

    return images[:n]
```

```python
# === ADK に SDXL Turbo 画像生成をツール登録する最小実装 ===
import os, time, random, torch
from io import BytesIO
from PIL import Image
from IPython.display import display
from diffusers import AutoPipelineForText2Image

# ADK
from google.adk.agents.llm_agent import LlmAgent
from google.adk.runners import Runner
from google.adk.sessions import InMemorySessionService

# すでに Gemini の TEXT_MODEL を使っているなら流用。未定義ならダミーでも可。
try:
    TEXT_MODEL  # noqa
except NameError:
    TEXT_MODEL = "gemini-2.5-flash"

# ---- SDXL Turbo パイプライン（遅延初期化 & 使い回し）----
_sdxl_pipe = None
_device = "cuda" if torch.cuda.is_available() else "cpu"
_dtype = torch.float16 if torch.cuda.is_available() else torch.float32

def _ensure_sdxl():
    global _sdxl_pipe
    if _sdxl_pipe is None:
        _sdxl_pipe = AutoPipelineForText2Image.from_pretrained(
            "stabilityai/sdxl-turbo", torch_dtype=_dtype
        ).to(_device)
    return _sdxl_pipe

# ---- ADK ツール：画像生成（Diffusers / 無料）----
def generate_image_sdxl(
    prompt: str,
    n: int = 1,
    steps: int = 4,
    guidance: float = 0.0,
    seed: int | None = 1234,
    display_in_colab: bool = True,
) -> dict:
    """
    SDXL-Turbo で画像を生成して保存パスを返すツール。
    - prompt: 英語の画像プロンプト
    - n: 枚数
    - steps: 反復ステップ（Turbo は 2〜6 で高速、品質↑なら 8〜12）
    - guidance: 0.0〜1.5 推奨（Turbo）
    - seed: 再現用。None なら都度ランダム
    - display_in_colab: True なら display() で表示
    return: {"paths": [...]}  生成画像の保存パス配列
    """
    pipe = _ensure_sdxl()

    paths = []
    base = f"/content/mascot_{int(time.time())}"
    for i in range(n):
        g = None
        if seed is not None:
            # 複数枚でも変化が出るようにオフセット
            g = torch.Generator(device=pipe.device).manual_seed(seed + i)

        out = pipe(
            prompt=prompt,
            num_inference_steps=steps,
            guidance_scale=guidance,
            generator=g,
        )
        img: Image.Image = out.images[0]
        path = f"{base}_{i+1}.png"
        img.save(path, "PNG")
        paths.append(path)
        if display_in_colab:
            display(img)

    return {"paths": paths}

# ---- ADK エージェント定義 ----
agent = LlmAgent(
    name="ImageGenAgent",
    instruction=(
        "You are an assistant that generates images using the SDXL Turbo tool. "
        "When the user provides a prompt (or mentions 'img_prompt'), "
        "call the tool 'generate_image_sdxl' with that prompt and return only the saved file paths."
    ),
    model=TEXT_MODEL,
    tools=[generate_image_sdxl],
)

runner = Runner(
   app_name="JokeImageApp",                # ← これが必須
    agent=agent,
    session_service=InMemorySessionService()
)
# ---- 実行ラッパ（img_prompt を自動で渡す）----
def adk_generate_from_img_prompt(
    prompt_var_name: str = "img_prompt",
    n: int = 1,
    steps: int = 4,
    guidance: float = 0.0,
    seed: int | None = 1234,
):
    if prompt_var_name not in globals():
        raise RuntimeError(f"{prompt_var_name} が見つかりません。先にプロンプトを作成してください。")
    p = globals()[prompt_var_name]

    msg = (
        "Use the tool 'generate_image_sdxl' with the following arguments and return only the file paths.\n"
        f"prompt: '''{p}'''\n"
        f"n: {n}\nsteps: {steps}\nguidance: {guidance}\nseed: {seed}\n"
    )

    # --- ADKの版差を吸収する呼び出し ---
    try:
        # まずは "prompt=" で
        result = runner.run(prompt=msg)
    except TypeError:
        try:
            # 一部版では "input_text=" / "query=" などの名前
            result = runner.run(input_text=msg)
        except TypeError:
            try:
                result = runner.run(query=msg)
            except TypeError:
                # どうしても合わない場合は、ツールを直接呼んで結果を返す（フォールバック）
                out = generate_image_sdxl(prompt=p, n=n, steps=steps, guidance=guidance, seed=seed)
                print(out["paths"])
                return

    print(getattr(result, "text", str(result)))
```

```python
# === 三択クイズ：画像 + ジョーク三択（正解=画像に使ったジョーク） ===
import os, random, yaml, requests, ipywidgets as w
from IPython.display import display, clear_output
from PIL import Image
from io import BytesIO

# Gemini (google-genai) を使う。未初期化なら config.yml から読む
try:
    client  # noqa
    TEXT_MODEL  # noqa
except NameError:
    from google import genai
    with open("config.yml", "r") as f:
        _cfg = yaml.safe_load(f)
    API_KEY    = _cfg["gemini_api_key"]
    TEXT_MODEL = _cfg.get("text_model", "gemini-2.5-flash")
    client     = genai.Client(api_key=API_KEY)

from google.genai import types

# ---- ユーティリティ ----
HEADERS = {"Accept": "application/json", "User-Agent": "Colab-quiz/1.0"}

def fetch_random_joke_en() -> str:
    r = requests.get("https://icanhazdadjoke.com/", headers=HEADERS, timeout=10)
    r.raise_for_status()
    return r.json().get("joke", "")

def translate_en_to_ja(text_en: str) -> str:
    """Geminiで英→日の自然な訳（1行）"""
    prompt = f"次の英語ジョークを自然な日本語に1行で翻訳してください。引用符は付けず、原文のニュアンスは保ってください。\n\n{text_en}"
    res = client.models.generate_content(
        model=TEXT_MODEL,
        contents=[
            types.Content(
                role="user",
                parts=[types.Part.from_text(text=prompt)]
            )
        ],
    )
    return (res.text or "").strip()

def pick_unique_distractors(correct_en: str, k: int = 2) -> list[str]:
    """正解と重複しないランダムジョークをk個集める"""
    s = set([correct_en.strip()])
    out = []
    for _ in range(20):  # 最大20回トライ
        if len(out) >= k:
            break
        j = fetch_random_joke_en().strip()
        if j and j not in s:
            s.add(j)
            out.append(j)
    if len(out) < k:
        # まれに重複続きの保険
        while len(out) < k:
            out.append("Why did the math book look sad? Because it had too many problems.")
    return out[:k]

# ---- 画像の取得（あなたのノートの変数から拾う）----
# 1) 既存の 'images' (PIL.Image のリスト) があればそれを優先
# 2) 次に 'paths' (保存ファイルパスのリスト) があれば最後の1枚を表示
# 3) どれも無ければスキップ
quiz_image = None
if "images" in globals() and images:
    quiz_image = images[0]
elif "paths" in globals() and paths:
    fp = paths[-1]
    if os.path.exists(fp):
        quiz_image = Image.open(fp)

# ---- 正解ジョーク（英語）は、あなたのノートの 'joke' 変数を利用。無ければ新規取得 ----
try:
    correct_en = joke  # 既存の英語ジョーク（画像の元）
except NameError:
    correct_en = fetch_random_joke_en()

correct_ja = translate_en_to_ja(correct_en)

# ---- 誤答（英語）をAPIから → 日本語に訳す ----
distractors_en = pick_unique_distractors(correct_en, k=2)
distractors_ja = [translate_en_to_ja(x) for x in distractors_en]

# ---- 三択データ作成（シャッフル）----
# ---- 三択データ作成（英語表示）----
# ここまでで correct_en, distractors_en が揃っている前提
options = [
    {"label": "A", "text": correct_en,        "is_correct": True},
    {"label": "B", "text": distractors_en[0], "is_correct": False},
    {"label": "C", "text": distractors_en[1], "is_correct": False},
]
random.shuffle(options)  # 並びをランダムに

# ---- UI ----
out = w.Output()

if 'quiz_image' in globals() and quiz_image is not None:
    display(quiz_image)

print("Which joke matches this image? (English options)")

rb = w.RadioButtons(
    options=[f"{opt['label']}. {opt['text']}" for opt in options],
    description="Pick:",
    layout=w.Layout(width="100%"),
)
btn = w.Button(description="Submit", button_style="primary")
show_ja = w.Checkbox(value=False, description="Show Japanese translation after grading")
display(rb, show_ja, btn, out)

def on_submit(_):
    with out:
        clear_output()
        selected_text = rb.value
        chosen = None
        for opt in options:
            if selected_text.startswith(opt["label"] + "."):
                chosen = opt
                break
        if chosen is None:
            print("Please choose an option.")
            return

        if chosen["is_correct"]:
            print("✅ Correct!")
        else:
            print("❌ Incorrect.")

        # 正解の英語を表示
        print("\n【Correct (EN)】", correct_en)

        # 任意：日本語訳も見たい場合（既に correct_ja があればそれを、無ければその場で翻訳）
        if show_ja.value:
            try:
                correct_ja  # 既に計算済みか？
            except NameError:
                # 翻訳関数が定義済みなら再利用、無ければスキップ
                if 'translate_en_to_ja' in globals():
                    _ja = translate_en_to_ja(correct_en)
                    print("【Correct (JA)】", _ja)
                else:
                    print("【Correct (JA)】 (translator function not found)")
        # （必要なら誤答の英語を表示する処理も追記可能）

btn.on_click(on_submit)

```

```python
import requests

HEADERS = {
    "Accept": "application/json",
    "User-Agent": "Colab-icanhaz-search/1.0"  # ヘッダはASCIIのみ
}

def search_jokes(term: str, limit: int = 10, page: int = 1):
    r = requests.get(
        "https://icanhazdadjoke.com/search",
        headers=HEADERS,
        params={"term": term, "limit": limit, "page": page},
        timeout=10,
    )
    r.raise_for_status()
    js = r.json()
    return [it["joke"] for it in js.get("results", [])], js

# 入力欄
term = input("検索語を入力してください（例: bird, halloween など）: ").strip()
if not term:
    term = "an"

jokes, meta = search_jokes(term, limit=10, page=1)

if jokes:
    print(f"▶ {term!r} の検索結果（{len(jokes)}件 / 全{meta.get('total_jokes', 0)}件中）")
    for i, j in enumerate(jokes, 1):
        print(f"{i}. {j}")
else:
    print("該当するジョークはありませんでした。")
```
